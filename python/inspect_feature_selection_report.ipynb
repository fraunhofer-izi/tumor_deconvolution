```python
import os
import pickle
import re
from multiprocessing import Pool, Manager
from tqdm.auto import tqdm
import pandas as pd
import numpy as np
from scipy.special import logsumexp
import plotnine as pn
```

```python
report_file = '/<censored_path>/dominik.otto/tumor-deconvolution-dream-challenge/feature_wse_report_fine.pkl'
#selection_out = '/<censored_path>/dominik.otto/tumor-deconvolution-dream-challenge/feature_whseq_corrected_selection_fine.txt'
with open(report_file, 'rb') as f:
    report = pickle.load(f)
```

```python
t_report = report.loc[report['feature'] == 'total', :]
mt_report = t_report.melt(id_vars=['iteration'], var_name='cell type', value_name='score')
bad_vars = set(['remove', 'index', 'signature', 'feature', 'nfeatures', 'total'])
ind = [var not in bad_vars for var in mt_report['cell type']]
mt_report = mt_report.loc[ind, :]
mt_report['score'] = mt_report['score'] .astype(np.float)
pn.ggplot(pn.aes('iteration', 'score', color='cell type'), mt_report) + pn.geom_line()
```

```python
pn.ggplot(pn.aes('iteration', 'score', color='cell type'), mt_report) + pn.geom_line() + pn.scale_y_log10(limits=[1e6, 1e9])
```

```python
total_factor = 7
t_report = report.loc[report['feature'] == 'total', :]
t_report['scaled total score'] = t_report['total'] / total_factor
mt_report = t_report.melt(id_vars=['iteration'])
good_vars = set(['nfeatures', f'scaled total score'])
ind = [var in good_vars for var in mt_report['variable']]
mt_report = mt_report.loc[ind, :]
mt_report['value'] = mt_report['value'] .astype(np.float)
pn.ggplot(pn.aes('iteration', 'value', color='variable'), mt_report) + pn.geom_line()
```
